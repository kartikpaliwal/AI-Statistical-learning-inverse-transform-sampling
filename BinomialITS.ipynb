```python
import numpy as np 
import matplotlib.pyplot as plt 
import math

N = 10
MinValue = 0
MaxValue = 10
P = 0.5

Frequency = [0]*((MaxValue - MinValue)+1)
RandomVariableValues = np.linspace(MinValue,MaxValue,((MaxValue-MinValue)+1))

def Factorial(Number):
    
    if Number == 0 or Number == 1:
        
        return 1
    else:
        
        return (Number * Factorial(Number-1))
    
    
def BinomialProbability(N,K):
    
    return ((Factorial(N)/(Factorial(K)*Factorial(N-K)))*(P**K)*((1-P)**(N-K)))


for i in range(0,10000):
    
    RandomProbability = np.random.uniform(0,1)
    CummulativeBinomialProbability = 0
    K = 0
    
    while (CummulativeBinomialProbability < RandomProbability):
        
        CummulativeBinomialProbability += BinomialProbability(N,K)
        K += 1
    
    K -= 1
    
    Frequency[K] += 1
```

```python
plt.plot(RandomVariableValues,Frequency)
plt.show()
```
